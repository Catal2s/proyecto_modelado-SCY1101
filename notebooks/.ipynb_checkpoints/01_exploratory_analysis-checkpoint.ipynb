Objetivo: Comprender la estructura, calidad y patrones de los datos de envíos logísticos.

## 1. Importación de librerías

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Configuración de visualizaciones
plt.rcParams['figure.figsize'] = (12, 5)
sns.set_style('whitegrid')
sns.set_palette('Set2')

import sys
sys.path.append('../src')
from data_preprocessing import (cargar_datos, limpiar_envios, limpiar_rutas,
                                  limpiar_vehiculos, construir_dataset_modelado)
print("Librerías cargadas correctamente.")
```

## 2. Carga de datos

```
envios_raw, incidencias_raw, rutas_raw, vehiculos_raw = cargar_datos()

print("=== ENVIOS ===")
print(f"Shape: {envios_raw.shape}")
display(envios_raw.head(3))

print("\n=== INCIDENCIAS ===")
print(f"Shape: {incidencias_raw.shape}")
display(incidencias_raw.head(3))

print("\n=== RUTAS ===")
print(f"Shape: {rutas_raw.shape}")
display(rutas_raw.head(3))

print("\n=== VEHÍCULOS ===")
print(f"Shape: {vehiculos_raw.shape}")
display(vehiculos_raw.head(3))
```

## 3. Calidad de datos — Valores nulos

```
fig, axes = plt.subplots(2, 2, figsize=(16, 10))
datasets = {'Envíos': envios_raw, 'Incidencias': incidencias_raw,
            'Rutas': rutas_raw, 'Vehículos': vehiculos_raw}

for ax, (nombre, df) in zip(axes.flatten(), datasets.items()):
    nulos = df.isnull().sum()
    nulos = nulos[nulos > 0]
    if len(nulos) > 0:
        nulos.plot(kind='bar', ax=ax, color='salmon')
        ax.set_title(f'Valores nulos — {nombre}')
        ax.set_ylabel('Cantidad')
        ax.set_xlabel('')
        ax.tick_params(axis='x', rotation=30)
    else:
        ax.text(0.5, 0.5, 'Sin nulos', ha='center', va='center', fontsize=14)
        ax.set_title(nombre)

plt.suptitle('Análisis de Valores Nulos por Dataset', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('../results/plots/nulos.png', dpi=150)
plt.show()
print("Porcentaje de nulos en envíos:")
print((envios_raw.isnull().sum() / len(envios_raw) * 100).round(2))
```

## 4. Limpieza y transformación de datos

```
envios    = limpiar_envios(envios_raw)
rutas     = limpiar_rutas(rutas_raw)
vehiculos = limpiar_vehiculos(vehiculos_raw)
df        = construir_dataset_modelado(envios, rutas, vehiculos, incidencias_raw)

print(f"Dataset final: {df.shape}")
print(f"\nDistribución variable objetivo:")
print(df['estado'].value_counts())
print(f"\nBalance binario (0=Entregado, 1=Problema):")
print(df['estado_binario'].value_counts())
```

## 5. Distribución de la variable objetivo

```
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Estados originales
df['estado'].value_counts().plot(kind='bar', ax=axes[0], color=sns.color_palette('Set2'))
axes[0].set_title('Distribución de Estados de Envío')
axes[0].set_ylabel('Cantidad')
axes[0].tick_params(axis='x', rotation=15)

# Binario
colores = ['#2ecc71', '#e74c3c']
df['estado_binario'].value_counts().plot(kind='pie', ax=axes[1],
    labels=['Entregado', 'Con Problema'], colors=colores,
    autopct='%1.1f%%', startangle=90)
axes[1].set_title('Variable Objetivo Binaria')
axes[1].set_ylabel('')

plt.suptitle('Variable Objetivo — Estado del Envío', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('../results/plots/target_distribucion.png', dpi=150)
plt.show()
```

## 6. Análisis de variables numéricas

```
num_cols = ['peso_kg', 'volumen_m3', 'dias_transito', 'distancia_km',
            'tiempo_estimado_hrs', 'peaje_total']

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
for ax, col in zip(axes.flatten(), num_cols):
    df[col].dropna().hist(bins=30, ax=ax, color='steelblue', edgecolor='white')
    ax.set_title(col)
    ax.set_ylabel('Frecuencia')

plt.suptitle('Distribución de Variables Numéricas', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('../results/plots/histogramas_numericas.png', dpi=150)
plt.show()
```

## 7. Variables numéricas vs. target (boxplots)

```
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
for ax, col in zip(axes.flatten(), num_cols):
    df.boxplot(column=col, by='estado_binario', ax=ax)
    ax.set_title(col)
    ax.set_xlabel('0=Entregado | 1=Problema')
    ax.set_ylabel(col)

plt.suptitle('Variables Numéricas por Estado del Envío', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('../results/plots/boxplots_vs_target.png', dpi=150)
plt.show()
```

## 8. Análisis de variables categóricas

```
cat_cols = ['tipo_carga', 'tipo_via', 'tipo']

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
for ax, col in zip(axes, cat_cols):
    if col in df.columns:
        tabla = pd.crosstab(df[col], df['estado_binario'], normalize='index') * 100
        tabla.plot(kind='bar', ax=ax, color=['#2ecc71','#e74c3c'], stacked=True)
        ax.set_title(f'{col} vs Estado')
        ax.set_ylabel('% de envíos')
        ax.legend(['Entregado', 'Problema'], loc='upper right')
        ax.tick_params(axis='x', rotation=20)

plt.suptitle('Variables Categóricas vs. Estado del Envío', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('../results/plots/categoricas_vs_target.png', dpi=150)
plt.show()
```

## 9. Matriz de correlación

```
corr_cols = ['peso_kg','volumen_m3','dias_transito','distancia_km',
              'tiempo_estimado_hrs','peaje_total','velocidad_media_kmh',
              'km_recorridos','antiguedad_años','tuvo_incidencia','estado_binario']

df_corr = df[corr_cols].dropna()
corr_matrix = df_corr.corr()

plt.figure(figsize=(13, 10))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm',
            mask=mask, vmin=-1, vmax=1, linewidths=0.5)
plt.title('Matriz de Correlación — Variables del Proyecto', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('../results/plots/correlacion.png', dpi=150)
plt.show()
```

## 10. Análisis de incidencias

```
print("Tipos de incidencias:")
print(incidencias_raw['tipo_incidencia'].value_counts())

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

incidencias_raw['tipo_incidencia'].value_counts().plot(
    kind='bar', ax=axes[0], color='coral')
axes[0].set_title('Frecuencia por Tipo de Incidencia')
axes[0].tick_params(axis='x', rotation=20)

incidencias_raw.boxplot(column='costo_impacto', by='tipo_incidencia', ax=axes[1])
axes[1].set_title('Costo de Impacto por Tipo de Incidencia')
axes[1].tick_params(axis='x', rotation=20)

plt.tight_layout()
plt.savefig('../results/plots/incidencias.png', dpi=150)
plt.show()
```

## 11. Resumen estadístico

```
print("=== ESTADÍSTICAS DESCRIPTIVAS ===")
display(df[num_cols + ['estado_binario']].describe().round(2))

print("\n=== INSIGHTS PRINCIPALES ===")
total = len(df)
pct_problema = df['estado_binario'].mean() * 100
print(f"• Total envíos: {total}")
print(f"• Envíos con problema: {pct_problema:.1f}%")
print(f"• Envíos con incidencias: {df['tuvo_incidencia'].mean()*100:.1f}%")
print(f"• Días de tránsito promedio: {df['dias_transito'].mean():.1f}")
print(f"• Distancia promedio: {df['distancia_km'].mean():.1f} km")
```
