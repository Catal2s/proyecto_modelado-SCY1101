# Notebook 04 — Optimización de Hiperparámetros
**SCY1101 Programación para la Ciencia de Datos**

Métodos: GridSearchCV · RandomizedSearchCV · Optuna

## 1. Setup

```
import sys, warnings
sys.path.append('../src')
warnings.filterwarnings('ignore')
import pandas as pd, numpy as np, matplotlib.pyplot as plt, seaborn as sns
from data_preprocessing import (cargar_datos, limpiar_envios, limpiar_rutas,
    limpiar_vehiculos, construir_dataset_modelado, preparar_features)
from hyperparameter_tuning import grid_search_rf, random_search_rf, optuna_search_rf
from model_evaluation import evaluar_modelo

envios, incidencias, rutas, vehiculos = cargar_datos()
envios    = limpiar_envios(envios)
rutas     = limpiar_rutas(rutas)
vehiculos = limpiar_vehiculos(vehiculos)
df = construir_dataset_modelado(envios, rutas, vehiculos, incidencias)
X_train, X_test, y_train, y_test, feature_names = preparar_features(df)
print("Setup completo.")
```

## 2. Baseline — Random Forest sin optimizar

```
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score

rf_base = RandomForestClassifier(n_estimators=100, random_state=42)
rf_base.fit(X_train, y_train)
f1_base = f1_score(y_test, rf_base.predict(X_test))
print(f"F1-Score baseline (sin optimizar): {f1_base:.4f}")
```

## 3. GridSearchCV
Búsqueda exhaustiva sobre una grilla predefinida de hiperparámetros.

```
rf_grid, df_grid = grid_search_rf(X_train, y_train, cv=5)
metricas_grid = evaluar_modelo(rf_grid, X_test, y_test, "RF GridSearch")
print("\nMétricas en Test Set — GridSearchCV:")
print(pd.Series(metricas_grid))
```

## 4. RandomizedSearchCV
Búsqueda aleatoria sobre un espacio de distribuciones más amplio.

```
rf_random, df_random = random_search_rf(X_train, y_train, n_iter=30, cv=5)
metricas_random = evaluar_modelo(rf_random, X_test, y_test, "RF RandomSearch")
print("\nMétricas en Test Set — RandomizedSearchCV:")
print(pd.Series(metricas_random))
```

## 5. Optuna
Optimización bayesiana: aprende del pasado para explorar mejor el espacio.

```
rf_optuna, study = optuna_search_rf(X_train, y_train, n_trials=50)
if rf_optuna is not None:
    metricas_optuna = evaluar_modelo(rf_optuna, X_test, y_test, "RF Optuna")
    print("\nMétricas en Test Set — Optuna:")
    print(pd.Series(metricas_optuna))
```

## 6. Comparación de los 3 métodos de optimización

```
resultados_opt = pd.DataFrame([
    {'Método': 'Baseline',          'F1-Score': f1_base,
     'Mejores params': 'n_estimators=100, defaults'},
    {'Método': 'GridSearchCV',      **{k:v for k,v in metricas_grid.items() if k!='Modelo'}},
    {'Método': 'RandomizedSearchCV',**{k:v for k,v in metricas_random.items() if k!='Modelo'}},
]).set_index('Método') if rf_optuna is None else pd.DataFrame([
    {'Método': 'Baseline',           'F1-Score': f1_base},
    {'Método': 'GridSearchCV',       'F1-Score': metricas_grid['F1-Score']},
    {'Método': 'RandomizedSearchCV', 'F1-Score': metricas_random['F1-Score']},
    {'Método': 'Optuna',             'F1-Score': metricas_optuna['F1-Score']},
]).set_index('Método')

display(resultados_opt)
resultados_opt.to_csv('../results/metrics/comparacion_optimizacion.csv')
```

## 7. Visualización del impacto de la optimización

```
f1_scores = {
    'Baseline':           f1_base,
    'GridSearchCV':       metricas_grid['F1-Score'],
    'RandomizedSearchCV': metricas_random['F1-Score'],
}
if rf_optuna is not None:
    f1_scores['Optuna'] = metricas_optuna['F1-Score']

plt.figure(figsize=(10, 5))
colores = ['#95a5a6','#3498db','#2ecc71','#e67e22']
bars = plt.bar(f1_scores.keys(), f1_scores.values(), color=colores[:len(f1_scores)])
for bar, val in zip(bars, f1_scores.values()):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.002,
             f'{val:.4f}', ha='center', va='bottom', fontsize=11, fontweight='bold')
plt.title('Impacto de la Optimización de Hiperparámetros en F1-Score')
plt.ylabel('F1-Score')
plt.ylim(0, max(f1_scores.values()) * 1.1)
plt.tight_layout()
plt.savefig('../results/plots/optimizacion_comparacion.png', dpi=150)
plt.show()
```

## 8. Historia de optimización Optuna

```
if study is not None:
    import optuna
    fig = optuna.visualization.matplotlib.plot_optimization_history(study)
    plt.title('Historia de Optimización — Optuna')
    plt.tight_layout()
    plt.savefig('../results/plots/optuna_history.png', dpi=150)
    plt.show()
    
    fig2 = optuna.visualization.matplotlib.plot_param_importances(study)
    plt.title('Importancia de Hiperparámetros — Optuna')
    plt.tight_layout()
    plt.savefig('../results/plots/optuna_param_importance.png', dpi=150)
    plt.show()
```

## 9. Conclusiones

```
print("""
CONCLUSIONES — Optimización de Hiperparámetros
===============================================
GridSearchCV:
  - Búsqueda exhaustiva, garantiza encontrar el mejor dentro de la grilla.
  - Más lento, ideal cuando el espacio es pequeño y se quiere certeza.

RandomizedSearchCV:
  - Explora más configuraciones en menos tiempo.
  - Bueno cuando el espacio de búsqueda es grande.

Optuna:
  - Optimización bayesiana: aprende qué zonas explorar.
  - Más eficiente que ambos anteriores en espacios continuos.
  - Ideal para producción y experimentación intensiva.
""")
```
